```python
import pandas as pd
marks=[12,23,34,45,56,6,778,89,9,90]
a=pd.Series(marks)
print(a)
#head
print(a.head(5))
print()
#tail
print(a.tail(2))
print()
#count
print(a.count())
print()
#sum
print(a.sum())
print()
#mean
print(a.mean())
print()
#median
print(a.median())
print()
#mode
print(a.mode())
print(a.describe())
```
